```python
import matplotlib.pyplot as plt

x=[3,8,5,15]
y=[5,13,1,21]
x1=[4,9,5,16]
y1=[4,12,0,20]

# creating x-axis
plt.xlabel('x')
plt.ylabel('y')
plt.plot(x,y)
plt.plot(x1,y1)
plt.title("Bachpan ki yaden")
plt.scatter(x,y,marker='*',s=120,label='water')
plt.scatter(x1,y1,marker='+',s=220,label='soil')
plt.legend(loc='lower right') # to show  
plt.grid()
plt.bar(x,y)
plt.show()

```
