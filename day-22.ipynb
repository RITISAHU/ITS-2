```python
import pandas as pd
import matplotlib.pyplot as plt

# URL of the dataset
url = 'https://raw.githubusercontent.com/redashu/Datasets/master/advertising.csv'

# Load the dataset from the URL
data = pd.read_csv(url)



# Scatter plot for TV , Sales
plt.subplot(3, 1, 1)  # 3 row, 1 columns, first subplot
plt.scatter(data['TV'], data['Sales'])
plt.title('TV and Sales')
plt.xlabel('TV')
plt.ylabel('Sales')

# Scatter plot for Radio , Sales
plt.subplot(3, 1, 2)  # 3 row, 1 columns, second subplot
plt.scatter(data['Radio'], data['Sales'])
plt.title('Radio and Sales')
plt.xlabel('Radio')
plt.ylabel('Sales')

# Scatter plot for Newspaper , Sales
plt.subplot(3, 1, 3)  # 3 row, 1 columns, third subplot
plt.scatter(data['Newspaper'], data['Sales'])
plt.title('Newspaper and Sales')
plt.xlabel('Newspaper')
plt.ylabel('Sales')


plt.show()
```
